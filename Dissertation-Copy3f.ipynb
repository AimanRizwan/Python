### KMeans: Find the Clusters

In this step, we will use k-means clustering to cluster our target variables, Speed per 50 meters. To do this, we will first fit these variables to the k-means algorithm and determine the best number of clusters. Determining the ideal number of clusters for our k-means model can be done by measuring the sum of the squared distances to the nearest cluster center aka inertia. Much like the scree plot in fig. 1 for PCA, the k-means scree plot below indicates the percentage of variance explained, but in slightly different terms, as a function of the number of clusters.

```python
x = df.iloc[:,20:60] #Selecting all features with speed
x.head(10) # Viewing first 10 rows
```

```python
scaled_X = StandardScaler().fit_transform(x) # Standardizing the Data
```

```python
# Using this loop to determine the best value of K i.e the number of clusters to be made. 
# The diagram is called the Elbow Method Diagram
ks = range(1, 10) # Setting range of values of k from 1-10
inertias = [] # The inertia is the sum of square distance empty list
for k in ks: # Initializing the loop
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(scaled_X)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o', color='black') # Ploting the values of k against inertia
plt.xlabel('number of clusters, k') # labeling x-axis
plt.ylabel('inertia') # Labeling y-axis
plt.xticks(ks) # Setting ticks equal to k value
plt.show() # Show the figure
```

Fig. 3 shows that after 2 clusters(at the elbow), the change in the value of inertia is no longer significant, and most likely, neither is the variance of the rest of the data after the elbow point. Therefore we can discard everything after k=2 and proceed to the last step in the process.

### Visualize and Interpret the Clusters

I did this project with a basic question in mind: can people be grouped based on rowing performance? Reducing all those features down to principal components and then visualizing the clusters in those principal components using k-means hints that the answer to my question is most likely yes.

```python
# finding the clusters based on input matrix "x"
model = KMeans(n_clusters = 2, init = "k-means++", max_iter = 300, n_init = 10, random_state = 0) # Initiating K-means model
y_clusters = model.fit_predict(scaled_X) # Fitting the data and creating clusters
x['clusters'] = y_clusters # assigning clusters to x dataframe
x.head(15) # Viewing the first 15 rows of x dataframe
```

```python
x_axis = df['speed_1100'] # Setting x-axis value
y_axis = df['speed_1300'] # Setting y-axis value
plt.figure(figsize = (10,8))  # Setting figure size
sns.scatterplot(x_axis,y_axis, hue = y_clusters,palette = ['r','g']) # Creating scatter plot using cluster hue
plt.title('Clusters by target variable', fontsize = 20) # Setting Title
plt.show() # Showing figure
```

### PCA : Reduce Dimensionality

In this step, we will find the optimal number of components which capture the greatest amount of variance in the data. In my case, as seen in Fig. 1 below, that number is three.

```python
x_pca = df.select_dtypes(include=np.number) # Selecting all numeric features to perform PCA
```

#### Standardize the data to have a mean of ~0 and a variance of 1

```python
X_std = StandardScaler().fit_transform(x_pca) # Standardizing the data
X_std # Showing Results
```

#### Create a PCA instance: pca

```python
pca = PCA() # Creating instance
principalComponents = pca.fit_transform(X_std) # Fitting data into instance
```

#### Plot the explained variances

```python
features = range(pca.n_components_) # Setting range of features
plt.figure(figsize = (20,8))  # Setting figure size
plt.bar(features, pca.explained_variance_ratio_, color='black') # Drawing bar plot of each feature
plt.xlabel('PCA features') # label x-axis
plt.ylabel('variance %') # Label y-axis
plt.xticks(features) # Set tick values equal to feature number
plt.show() # Show figure
```

The two figures show that the first three components contain the largest eigen values which explain the majority of the variance in our data. For this visualization use case, we will quickly plot just the first two. We do this to notice if there are any clear clusters.

```python
pca = PCA(3)  # Initialize PCA to create 3 components only
principalComponents = pca.fit_transform(X_std) #Fitting datainto PCA
```

#### Save components to a DataFrame

```python
PCA_components = pd.DataFrame(principalComponents) #saving components to dataframe
PCA_components['labels'] =  y_clusters # assigning clusters to new dataframe
PCA_components # Viewing the dataframe
```

###  Classification Analysis using PCA + Kmeans Dataset

#### Creating a copy of reduced dataset

```python
final_df = PCA_components.copy() # Creating a copy of finalized dataframe
final_df.head() # Viewing first 5 rows
```

```python
# Renaming the column names
final_df = final_df.rename(columns={final_df.columns[0]: 'Component 1', final_df.columns[1]: 'Component 2',final_df.columns[2]: 'Component 3' })
final_df
```

#### Visualizing the Correlation of Components with Clusters

#### Splitting Data into Train/Test

```python
# split dataset in features and target variable
X = final_df[['Component 1', 'Component 2','Component 3']] # Predictor Variable    
y = final_df.labels # Target variable
```

```python
# Train/Test Split
# 70% training and 30% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify = y) 
```

#### Creating Classifier Models Pipeline

In order to create a pipeline, I append the default state of all classification algorithms mentioned above into the model list and then iterate through them to train, test, predict and evaluate.

```python
model_pipeline = [] # Empty pipeline list
model_pipeline.append(LogisticRegression(solver='liblinear')) # Appending Logistic regression to pipeline
model_pipeline.append(SVC()) # Appending SVC to pipeline
model_pipeline.append(KNeighborsClassifier()) # Appending KNN to pipeline
model_pipeline.append(DecisionTreeClassifier()) # Appending Decision Tree to pipeline
model_pipeline.append(RandomForestClassifier()) # Appending  Random Forest to pipeline
model_pipeline.append(GaussianNB()) # Appending Naive Bayes to pipeline
```

```python
from sklearn.metrics import precision_score
```

```python
model_list = ['Logistic Regression', 'SVM', 'KNN', 'Decision Tree', 'Random Forest', 'Naive Bayes'] # Creating model list
acc_list = [] # Accuracy Empty List
auc_list = [] # AUC Empty List
cm_list = [] # Confusion Matrix Empty List
precision_list = [] #Precision Empty List
recall_list = [] #Recall Empty List

for model in model_pipeline: # Initializing Loop
    model.fit(X_train, y_train) # Fitting each Algorithm one by one
    y_pred = model.predict(X_test) # Predicing values of each algorithm ony by one
    acc_list.append(metrics.accuracy_score(y_test, y_pred)) # Appending accuracy of each model to acc_list
    fpr, tpr, _thresholds = metrics.roc_curve(y_test, y_pred,pos_label=1) # calculating roc of each model to find auc
    auc_list.append((round(metrics.auc(fpr, tpr),2))) # Appending AUC of each model to auc_list
    cm_list.append(confusion_matrix(y_test, y_pred)) # Appending CM of each model to cm_list
    precision_list.append(metrics.precision_score(y_test, y_pred)) # Appending precision of each model to precision_list
    recall_list.append(metrics.recall_score(y_test, y_pred)) # Appending recall_score of each model to recall_list
```

### Comparing Confusion Matrix of Classifiers

Confusion matrix indicates the actual values vs. predicted values and summarize the true negative, false positive, false negative and true positive values in a matrix format. It Basically shows How many values the model predicted correctly and how many values the trained model predicted wrong. True Positve = (Value Correctly Predicted by ML Model), True Negative = (Wrongly Predicted by ML Model)

```python
## plot confuston matrix
fig = plt.figure(figsize = (24,10)) # setting figure size
for i in range(len(cm_list)): # Initializing the loop
    cm = cm_list[i] # Assigning CM to each model per loop
    model = model_list[i] # Assigning model per loop
    sub = fig.add_subplot(2, 3, i+1).set_title(model) # Creating figure dimensions
    cm_plot = sns.heatmap(cm, annot=True, cmap = 'Blues_r', linewidths=0.4, annot_kws={"size":12}, fmt='g') # Drawing CM
    cm_plot.set_xlabel('Predicted Values') # Labelling x-axis
    cm_plot.set_ylabel('Actual Values')  # Labeling y-axis
```

### Comparing AUC and Accuracy of the Models

Accuracy is the most straightforward indicator of the model performance. It measure the percentage of accurate predictions: accuracy = (true positive + true negative) / (true positive + false positive + false negative + false positive)+

ROC is the plot of true positive rate against false positive rate at various classification threshold. AUC is the area under the ROC curve, and higher AUC indicates better model performance.

```python
# accuracy and AUC
result_df = pd.DataFrame({'Model' :model_list, 'Accuracy': acc_list, 'AUC': auc_list,
                         'Precision' :precision_list,'Recall' :recall_list}) # Creating a dataframe of the results
result_df # Viewing the results
```

```python
# Adding mean and standard deviation of each classification.
result_df["Mean"] = result_df.iloc[:,1:].mean(axis = 1)
result_df["Median"] = result_df.iloc[:,1:].median(axis = 1)
result_df["Std"] = result_df.iloc[:,1:].std(axis = 1)
```
